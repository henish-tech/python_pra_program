```python
print("Welcome to the Smart Data Processor!\n")


def show_menu():
    print("1. Enter Data")
    print("2. Display Data Summary")
    print("3. Calculate Factorial (Recursion)")
    print("4. Filter Data Above Threshold (Lambda)")
    print("5. Sort Data (Ascending/Descending)")
    print("6. Get Data Statistics (Multiple Returns)")
    print("7. Exit Program\n")

def get_data():
    data = input("Enter numbers separated by spaces: ")
    return list(map(int, data.split()))


def data_summary(data):
    print("\n--- Data Summary ---")
    print("Total elements:", len(data))
    print("Minimum value:", min(data))
    print("Maximum value:", max(data))
    print("Sum of values:", sum(data))
    print("Average value:", sum(data)/len(data))
    print()


def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n - 1)


def filter_threshold(data, threshold):
    result = list(filter(lambda x: x >= threshold, data))
    return result

def sort_data(data, order):
    if order == '1':
        sorted_data = sorted(data)
        print("\nData sorted in ascending order:")
    elif order == '2':
        sorted_data = sorted(data, reverse=True)
        print("\nData sorted in descending order:")
    else:
        print("Invalid sort option.")
        return []
    print(sorted_data)
    return sorted_data


def get_stats(data):
    total = sum(data)
    avg = total / len(data)
    smallest = min(data)
    largest = max(data)
    return smallest, largest, total, avg


def exit_program():
    print("\nThank you for using the Smart Data Processor!")
    print("Goodbye!\n")


show_menu()

data = []

while True:
    choice = input("Enter your choice: ")

    if choice == '1':
        data = get_data()
        print("\nData stored successfully!\n")

    elif choice == '2':
        if data:
            data_summary(data)
        else:
            print("No data available. Please enter data first.\n")

    elif choice == '3':
        num = int(input("Enter a number to find its factorial: "))
        print("Factorial of", num, "is:", factorial(num))
        print()

    elif choice == '4':
        if data:
            threshold = int(input("Enter threshold value: "))
            result = filter_threshold(data, threshold)
            print(f"Filtered Data (>= {threshold}): {result}\n")
        else:
            print("Please enter data first.\n")

    elif choice == '5':
        if data:
            print("\n1. Ascending")
            print("2. Descending")
            order = input("Choose sorting order: ")
            sort_data(data, order)
            print()
        else:
            print("Please enter data first.\n")

    elif choice == '6':
        if data:
            small, large, total, avg = get_stats(data)
            print("\n--- Data Statistics ---")
            print("Smallest value:", small)
            print("Largest value:", large)
            print("Sum of values:", total)
            print("Average value:", avg)
            print()
        else:
            print("Please enter data first.\n")

    elif choice == '7':
        exit_program()
        break

    else:
        print("Invalid choice! Please try again.\n")

    show_menu()
```
