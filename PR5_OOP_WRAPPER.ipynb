```python
print('--- Python OOP Project: Student Management System ---\n')

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def show_details(self):
        print(f"Name: {self.name}, Age: {self.age}")



class Student(Person):
    def __init__(self, name, age, student_id, grade):
        super().__init__(name, age)
        self.student_id = student_id
        self.grade = grade

    def show_details(self):
        super().show_details()
        print(f"Student ID: {self.student_id}, Grade: {self.grade}")



class ClassRepresentative(Student):
    def __init__(self, name, age, student_id, grade, responsibility):
        super().__init__(name, age, student_id, grade)
        self.responsibility = responsibility

    def show_details(self):
        super().show_details()
        print(f"Responsibility: {self.responsibility}")

people = []
students = {}
representatives = {}


def compare_grades(id1, id2):
    student1 = students.get(id1) or representatives.get(id1)
    student2 = students.get(id2) or representatives.get(id2)

    if not student1 or not student2:
        print("Invalid student IDs.\n")
        return

    print("\n--- Comparing Grades ---")
    if student1.grade > student2.grade:
        print(f"{student1.name} ({student1.student_id}) has a higher grade than {student2.name} ({student2.student_id}).")
    elif student1.grade < student2.grade:
        print(f"{student1.name} ({student1.student_id}) has a lower grade than {student2.name} ({student2.student_id}).")
    else:
        print(f"{student1.name} and {student2.name} have the same grade.")
    print()



while True:
    print("\nChoose an operation:")
    print("1. Create a Person")
    print("2. Create a Student")
    print("3. Create a Class Representative")
    print("4. Show Details")
    print("5. Compare Grades")
    print("6. Exit\n")

    choice = input("Enter your choice: ")

  
    if choice == '1':
        name = input("Enter name: ")
        age = input("Enter age: ")
        person = Person(name, age)
        people.append(person)
        print(f"Person created successfully: {name}, Age: {age}")

   
    elif choice == '2':
        name = input("Enter name: ")
        age = input("Enter age: ")
        student_id = input("Enter student ID: ")
        grade = input("Enter grade (A/B/C): ")
        student = Student(name, age, student_id, grade)
        students[student_id] = student
        print(f"Student created successfully: {name} ({student_id})")

    
    elif choice == '3':
        name = input("Enter name: ")
        age = input("Enter age: ")
        student_id = input("Enter student ID: ")
        grade = input("Enter grade (A/B/C): ")
        responsibility = input("Enter responsibility (e.g., Discipline, Events): ")
        rep = ClassRepresentative(name, age, student_id, grade, responsibility)
        representatives[student_id] = rep
        print(f"Class Representative created successfully: {name} ({student_id})")

   
    elif choice == '4':
        print("\nSelect what to display:")
        print("1. All Persons")
        print("2. Specific Student")
        print("3. Specific Class Representative")

        sub_choice = input("Enter your choice: ")

        if sub_choice == '1':
            print("\n--- Person Details ---")
            for p in people:
                p.show_details()

        elif sub_choice == '2':
            student_id = input("Enter Student ID: ")
            if student_id in students:
                print("\n--- Student Details ---")
                students[student_id].show_details()
            else:
                print("Student not found.")

        elif sub_choice == '3':
            student_id = input("Enter Representative ID: ")
            if student_id in representatives:
                print("\n--- Class Representative Details ---")
                representatives[student_id].show_details()
            else:
                print("Representative not found.")
        else:
            print("Invalid option.")

   
    elif choice == '5':
        print("\nEnter two student IDs to compare grades.")
        id1 = input("Enter first student ID: ")
        id2 = input("Enter second student ID: ")
        compare_grades(id1, id2)

    elif choice == '6':
        print("\nExiting the system...")
        print("Thank you for using the Student Management System!")
        break

    else:
        print("Invalid choice. Please try again.")
```
