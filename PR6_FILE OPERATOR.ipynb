```python
import datetime
import os

print("Welcome to Personal Diary System!")

class Diary:
    def __init__(self, filepath="diary.txt"):
        self.filepath = filepath

    # Add a new diary entry
    def write_entry(self):
        entry_text = input("Type your diary entry: ")
        timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        entry = f"{timestamp} -> {entry_text}\n"
        try:
            with open(self.filepath, "a") as file:
                file.write(entry)
            print("Entry saved successfully!")
        except Exception as e:
            print("Error writing entry:", e)

    # View all entries
    def read_entries(self):
        if not os.path.exists(self.filepath):
            print("Diary is empty.")
            return
        with open(self.filepath, "r") as file:
            lines = file.readlines()
            if lines:
                print("\n--- Diary Entries ---")
                for line in lines:
                    print(line.strip())
            else:
                print("Diary is empty.")

    # Search entries by keyword
    def find_entry(self):
        keyword = input("Enter keyword to search: ").lower()
        if not os.path.exists(self.filepath):
            print("Diary file not found.")
            return
        with open(self.filepath, "r") as file:
            found = [line.strip() for line in file if keyword in line.lower()]
        if found:
            print("\n--- Search Results ---")
            for f in found:
                print(f)
        else:
            print(f"No entries found containing '{keyword}'.")

    # Delete all entries
    def clear_diary(self):
        if os.path.exists(self.filepath):
            open(self.filepath, "w").close()
            print("All entries deleted successfully!")
        else:
            print("Diary file does not exist.")


def show_menu():
    print("\nSelect an option:")
    print("1. Add a new entry")
    print("2. View all entries")
    print("3. Search for an entry")
    print("4. Delete all entries")
    print("5. Exit")


def main():
    diary = Diary()

    while True:
        show_menu()
        choice = input("Enter your choice (1-5): ")

        if choice == "1":
            diary.write_entry()
        elif choice == "2":
            diary.read_entries()
        elif choice == "3":
            diary.find_entry()
        elif choice == "4":
            diary.clear_diary()
        elif choice == "5":
            print("Exiting Personal Diary System. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")


if __name__ == "__main__":
    main()
```
