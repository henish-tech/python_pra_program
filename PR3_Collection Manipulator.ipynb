```python
def display_welcome():
    print("\nWelcome to the Classroom Data Manager!\n")
    print("This system helps organize both students and teachers' data.\n")

def display_menu():
    print("Select an option:")
    print("1. Add Student")
    print("2. Add Teacher")
    print("3. Display All Records")
    print("4. Update Student Details")
    print("5. Delete Student")
    print("6. Display Subjects Offered")
    print("7. Exit\n")

# Initialize lists
students = [5]
teachers = [10]

def add_student():
    print("\n--- Add New Student ---")
    sid = input("Enter student ID: ")
    name = input("Enter student name: ")
    age = input("Enter student age: ")
    grade = input("Enter grade: ")
    subjects = input("Enter subjects (comma separated): ").split(',')

    students.append({
        'id': sid,
        'name': name,
        'age': age,
        'grade': grade,
        'subjects': [s.strip() for s in subjects]
    })
    print("Student added successfully!\n")

def add_teacher():
    print("\n--- Add New Teacher ---")
    tid = input("Enter teacher ID: ")
    name = input("Enter teacher name: ")
    subject = input("Enter subject taught: ")
    exp = input("Enter years of experience: ")

    teachers.append({
        'id': tid,
        'name': name,
        'subject': subject,
        'experience': exp
    })
    print("Teacher added successfully!\n")

def display_all_records():
    print("\n--- All Student Records ---")
    if not students:
        print("No students available.")
    else:
        for s in students:
            subjects = ", ".join(s['subjects'])
            print(f"ID: {s['id']}, Name: {s['name']}, Age: {s['age']}, Grade: {s['grade']}, Subjects: {subjects}")

    print("\n--- All Teacher Records ---")
    if not teachers:
        print("No teachers available.\n")
    else:
        for t in teachers:
            print(f"ID: {t['id']}, Name: {t['name']}, Subject: {t['subject']}, Experience: {t['experience']} years")
    print()

def update_student():
    sid = input("\nEnter student ID to update: ")
    for s in students:
        if s['id'] == sid:
            print("Select detail to update:")
            print("1. Age")
            print("2. Grade")
            print("3. Subjects")
            choice = input("Choice: ")

            if choice == '1':
                s['age'] = input("Enter new age: ")
                print("Age updated successfully!")
            elif choice == '2':
                s['grade'] = input("Enter new grade: ")
                print("Grade updated successfully!")
            elif choice == '3':
                new_subjects = input("Enter new subjects (comma separated): ")
                s['subjects'] = [sub.strip() for sub in new_subjects.split(',')]
                print("Subjects updated successfully!")
            else:
                print("Invalid choice.")
            return
    print("Student not found.\n")

def delete_student():
    sid = input("\nEnter student ID to delete: ")
    for i, s in enumerate(students):
        if s['id'] == sid:
            del students[i]
            print("Student deleted successfully!\n")
            return
    print("Student not found.\n")

def display_subjects():
    subject_set = set()
    for s in students:
        subject_set.update(s['subjects'])
    if subject_set:
        print("\nSubjects Offered: " + ", ".join(subject_set) + "\n")
    else:
        print("\nNo subjects to display.\n")

def exit_program():
    print("\nThank you for using the Classroom Data Manager!\n")

def main():
    display_welcome()
    while True:
        display_menu()
        choice = input("Enter your choice: ")

        if choice == '1':
            add_student()
        elif choice == '2':
            add_teacher()
        elif choice == '3':
            display_all_records()
        elif choice == '4':
            update_student()
        elif choice == '5':
            delete_student()
        elif choice == '6':
            display_subjects()
        elif choice == '7':
            exit_program()
            break
        else:
            print("Invalid choice. Try again.\n")

if __name__ == "__main__":
    main()
```
