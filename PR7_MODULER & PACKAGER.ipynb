```python
def create_file(filename):
    try:
        with open(filename, "x") as f:
            print(f"File '{filename}' created successfully.")
    except FileExistsError:
        print(f"File '{filename}' already exists.")


def write_file(filename):
    content = input("Enter content to write -> ")
    with open(filename, "w") as f:
        f.write(content + "\n")
    print(f"Content written to '{filename}' successfully.")


def read_file(filename):
    try:
        with open(filename, "r") as f:
            data = f.read()
            print(f"\n--- Content of '{filename}' ---")
            print(data)
            print("-----------------------------")
    except FileNotFoundError:
        print(f"File '{filename}' not found!")


def append_file(filename):
    content = input("Enter content to append -> ")
    with open(filename, "a") as f:
        f.write(content + "\n")
    print(f"Content appended to '{filename}' successfully.")
```

```python
from datetime import datetime
import time
import math
import random
import string
import uuid
import file

print("welcome to ...")

while True:
    print("\nChoose an option -> ")
    print("1.Datetime and Time Operations")
    print("2.Mathematical Operations")
    print("3.Random Data Generation")
    print("4.Generate Unique Identifiers")
    print("5.File Operation (Custom Module")
    print("6.Explore Module Attributes (dir())")
    print("7.Exit")

    c = int(input("Enter the Choice -> "))

    if c == 1:
        while True:
            print("\nDatetime And Time Operations : ")
            print("1.Display Current Date And Time")
            print("2.Calculate Difference Between Two Dates/times")
            print("3.Format date into custom format")
            print("4.Stopwatch")
            print("5.Countdown Timer")
            print("6.Back to Main Menu")
            d = int(input("Enter the Choice -> "))
            if d == 1:
                print("Current time -> ", datetime.now())
                print("-" * 25)
            elif d == 2:
                d1 = input("Enter Date-1 (YYYY-MM-DD) -> ")
                d2 = input("Enter Date-2 (YYYY-MM-DD) -> ")
                date1=datetime.strptime(d1,"%Y-%m-%d")
                date2=datetime.strptime(d2,"%Y-%m-%d")
                dif = abs((date2 - date1).days)
                print("Date-1 => ", date1)
                print("Date-2 => ", date2)
                print(f"Difference Between Two days -> {dif}days")
                print("-" * 25)
            elif d == 3:
                date = input("Enter Date Formate (YYYY-MM-DD) ->")
                n_date = datetime.strptime(date, "%Y-%m-%d")
                new_date=n_date.strftime("%d-%m-%Y")
                print(f"Formated Date into Custom Format -> {new_date}")
                print("-" * 25)
            elif d == 4:
                print("Press Enter to Start Stopwatch")
                input()
                start_time = time.time()
                print("Press Enter to Stop Stopwatch")
                input()
                end_time = time.time()
                s = end_time - start_time
                print("Elapsed Time -> ", round(s, 2), "second")
                print("-" * 25)
            elif d == 5:
                s = input("Enter the Second (HH:MM:SS) ->")
                while True:
                    c = datetime.now().strftime("%H:%M:%S")
                    if c == s:
                        print("Time Completed....!")
                        break
                    time.sleep(1)
                print("-" * 25)
            elif d == 6:
                print("Thank You ...!")
                break
            else:
                print("Invalid Choice...!")

    elif c == 2:
        while True:
            print("\nMathematical Operations : ")
            print("1.Calculate Factorial")
            print("2.Solve Compound Interest")
            print("3.Trigonometric Calculations")
            print("4.Area Of Geometric Shapes")
            print("5.Back to main Menu")
            m = int(input("Enter The Choice -> "))
            if m == 1:
                n = int(input('\nEnter Number to Calculate Factorial -> '))
                fa = math.factorial(n)
                print(f"The Factorial of {n} -> {fa}")
                print("-"*25)
            elif m == 2:
                amount = float(input("\nEnter Principal amount -> "))
                interest = float(input("Enter the Interest -> "))
                year = int(input("Enter the Time In year -> "))
                result = amount * (1 + interest / 100) ** year
                print(f"Compound Interest -> {result}")
                print("-" * 25)
            elif m == 3:
                degree = int(input("\nEnter the degree -> "))
                radius = math.radians(degree)
                print(f"sin({degree}) -> {math.sin(radius)}")
                print(f"cos({degree}) -> {math.cos(radius)}")
                print(f"tan({degree}) -> {math.tan(radius)}")
                print("-" * 25)
            elif m == 4:
                print("\nChoose shape:")
                print("1. Square")
                print("2. Rectangle")
                print("3. Triangle")
                print("4. Circle")
                shape_choice = input("Enter your choice: ")
                if shape_choice == '1':
                    side = float(input("Enter the Side Of Square -> "))
                    print(f"Area of Square -> {side*side}")
                    print("-" * 25)
                elif shape_choice == '2':
                    length = float(input("Enter the length -> "))
                    width = float(input("Enter the width -> "))
                    print("Area of Rectangle -> ", length * width)
                    print("-" * 25)
                elif shape_choice == '3':
                    base = float(input("Enter the base -> "))
                    height = float(input("Enter the height -> "))
                    print("Area of Triangle -> ", 0.5 * base * height)
                    print("-" * 25)
                elif shape_choice == '4':
                    radius = float(input("Enter the Radius -> "))
                    print("Area Of Circle -> ", math.pi * radius * radius)
                    print("-" * 25)
                else:
                    print("Invalid Choice...!")
            elif m == 5:
                break
            else:
                print("Invalid Choice ..!")
    elif c == 3:
        while True:
            print("\nRandom Data Generation : ")
            print("1.Generate Random Number")
            print("2.Generate Random List")
            print("3.Create Random Password")
            print("4.Generate Random OTP")
            print("5.Back To main menu")
            r = int(input("Enter the choice -> "))
            if r == 1:
                num = random.randint(1,100)
                print("Random Number is -> ", num)
                print("-" * 25)
            elif r == 2:
                fruit = ['apple', 'mango', 'orange', 'banana', 'cherry']
                print(f"Random fruit from List -> {random.choice(fruit)}")
                print("-" * 25)
            elif r == 3:
                length = int(input("Enter the Length og Password -> "))
                character = string.ascii_letters + string.digits + '@' + string.punctuation
                password = ''
                for i in range(length):
                    password += random.choice(character)
                print("Ramdom Password -> ",password)
                print("-" * 25)
            elif r == 4:
                otp = ''
                for i in range(6):
                    otp += str(random.randint(0, 9))
                print(f"Your OTP -> {otp}")
                print("-" * 25)
            elif r == 5:
                break
            else:
                print("Invalid Choice ...!")
    elif c == 4:
        print("\nGenerate Unique Identifiers -> ")
        print(f"\nGenerate UUID -> {uuid.uuid4()} ")

    elif c == 5:
        while True:
            print("\n--- File Operation ---")
            print("1.Create File")
            print("2.Write to a File")
            print("3.Read from a File")
            print("4.Append to a File")
            print("5.Back to main menu")
            f = int(input("Enter the choice -> "))
            if f == 1:
                file_name = input("Enter File Name -> ")
                file.create_file(file_name)
                print("-" * 25)
            elif f == 2:
                file_name = input("Enter File Name -> ")
                file.write_file(file_name)
                print("-" * 25)
            elif f == 3:
                file_name = input("Enter File Name -> ")
                file.read_file(file_name)
                print("-" * 25)
            elif f == 4:
                file_name = input("Enter File Name -> ")
                file.append_file(file_name)
                print("-" * 25)
            elif f == 5:
                break
            else:
                print("Invalid Choice...!")
    elif c == 6:
        print("\n-- Explore Module Attributes ---")
        module = input("Enter module name to explore -> ")
        print(f"Available Attributes in {module} module -> {dir(module)}")
        print("-" * 25)
    elif c == 7:
        print("\nThank You Visit Again....!")
        break
    else:
        print("Invalid Choice...!")

print("------------------------------------------------------------------------------------------------")
```
