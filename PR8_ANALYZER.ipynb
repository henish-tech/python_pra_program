```python
import numpy as np

def get_array_input(size):
    elements = input(f"Enter {size} numbers separated by space: ").split()
    return np.array(list(map(int, elements)))

def create_array():
    print("\nSelect array type:")
    print("1. 1D Array\n2. 2D Array\n3. 3D Array")
    choice = int(input("Choice: "))

    if choice == 1:
        n = int(input("Number of elements: "))
        arr = get_array_input(n)

    elif choice == 2:
        rows = int(input("Number of rows: "))
        cols = int(input("Number of columns: "))
        arr = get_array_input(rows*cols).reshape(rows, cols)

    elif choice == 3:
        x = int(input("Size of first dimension: "))
        y = int(input("Size of second dimension: "))
        z = int(input("Size of third dimension: "))
        arr = get_array_input(x*y*z).reshape(x, y, z)

    else:
        print("Invalid choice!")
        return None

    print("Array created:\n", arr)
    return arr

def mathematical_operations(arr):
    print("\nMathematical Operations:\n1.Add\n2.Subtract\n3.Multiply\n4.Divide")
    choice = int(input("Choice: "))
    other = get_array_input(arr.size).reshape(arr.shape)

    operations = {
        1: ("Addition", np.add),
        2: ("Subtraction", np.subtract),
        3: ("Multiplication", np.multiply),
        4: ("Division", np.divide)
    }

    if choice in operations:
        name, func = operations[choice]
        print(f"{name} Result:\n{func(arr, other)}")
    else:
        print("Invalid choice!")

def combine_or_split(arr):
    print("\n1. Combine with another array\n2. Split horizontally")
    choice = int(input("Choice: "))
    other = get_array_input(arr.size).reshape(arr.shape)

    if choice == 1:
        combined = np.vstack((arr, other))
        print("Combined Array:\n", combined)
    elif choice == 2:
        if arr.ndim < 2:
            print("Cannot split 1D array horizontally!")
            return
        splits = np.hsplit(arr, arr.shape[1])
        print("Split Arrays:")
        for i, s in enumerate(splits):
            print(f"Part {i+1}:\n{s}")
    else:
        print("Invalid choice!")

def search_sort_filter(arr):
    print("\n1. Search Value\n2. Sort Array\n3. Filter Values > threshold")
    choice = int(input("Choice: "))

    if choice == 1:
        val = int(input("Value to search: "))
        print("Indices:", np.where(arr == val))
    elif choice == 2:
        sorted_arr = np.sort(arr, axis=None if arr.ndim==1 else 1)
        print("Sorted Array:\n", sorted_arr)
    elif choice == 3:
        threshold = int(input("Threshold: "))
        filtered = arr[arr > threshold]
        print(f"Filtered values: {filtered}")
    else:
        print("Invalid choice!")

def aggregates_statistics(arr):
    print("\nStatistics:\n1.Sum\n2.Mean\n3.Median\n4.Std Dev\n5.Variance")
    choice = int(input("Choice: "))

    stats = {
        1: np.sum,
        2: np.mean,
        3: np.median,
        4: np.std,
        5: np.var
    }

    if choice in stats:
        print("Result:", stats[choice](arr))
    else:
        print("Invalid choice!")

def main():
    arr = None
    while True:
        print("\n=== NumPy Toolkit ===")
        print("1.Create Array\n2.Math Operations\n3.Combine/Split\n4.Search/Sort/Filter\n5.Statistics\n6.Exit")
        choice = int(input("Select option: "))

        if choice == 1:
            arr = create_array()
        elif choice == 2 and arr is not None:
            mathematical_operations(arr)
        elif choice == 3 and arr is not None:
            combine_or_split(arr)
        elif choice == 4 and arr is not None:
            search_sort_filter(arr)
        elif choice == 5 and arr is not None:
            aggregates_statistics(arr)
        elif choice == 6:
            print("Exiting program. Goodbye!")
            break
        else:
            print("Invalid choice or array not created!")

if __name__ == "__main__":
    main()
```
